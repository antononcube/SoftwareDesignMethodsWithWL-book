## Setup

```raku
use Data::Importers;
use Markdown::Grammar;
use JSON::Fast;
use LLM::Configurations;
```

## Ingest Markdown dictionary

```raku
my $fileName = $*CWD ~ '/Design-Patterns-WL-Mermaid-JS.md';
my $mtext = slurp($fileName);

text-stats($mtext)
```

```raku
my @tree = md-section-tree($mtext);
```

```raku
@tree.head.value
```

```raku
#% html
@tree.head.values ==> to-html
```

```raku
my @dsCode = @tree.head.value.grep({ $_.key.defined && $_.value.defined && $_.elems }).map({ $_.key => $_.value.grep({ $_.<content>.trim.chars }).Array });;

deduce-type(@dsCode):tally
```

```raku
#% html

@dsCode ==> to-html
```

```raku
my %genRules = @dsCode.map({ $_.value.tail<content> => $_.value.head<content> })
```

-----

## Prompt

```raku
my $prompt = q:to/END/;
You are a translator natural language specifications or UML specifications of Object Oriented Programming designs into Mathematica code.

Here are example translations:
END

$prompt = $prompt ~ "\n\n" ~ %genRules.map({ "Input:\n{$_.key}\n\nOutput:\n{$_.value}"})
```

## Experiments

### Complicated prompt

```raku
my $res = llm-synthesize([
    $prompt,
    "Write WL code a Template method application for family of algorithms for matrix alorithms for:",
    "- transposing itself",
    "- dot product, and elementwise addition and multiplication with another matrix",
    "Have a concrete class for dense matrices and a concrete class for sparse matrices.",
    "Make an example template method operation, `SpecialOp`, that uses the other operations.",
    "Do not generate the full code, just the class structure.",
    llm-prompt('NothingElse')('Mathematica')
    ],
    e => $conf4o
)
```

### Examples function

```raku
my &toWLDP = llm-example-function(%genRules, e => $conf4o)
```
